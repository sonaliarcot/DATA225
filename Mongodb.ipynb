```python
from pymongo import MongoClient 
import pandas as pd
import json
from pymongo import MongoClient, DESCENDING
from pprint import pprint
```

# Setting up the connection

```python
client = MongoClient() 
```

```python
client = MongoClient('localhost',27017) 
```

```python
db = client.twitter
```

# Uplaoding CSV files into mongodb compass

```python
colname = db.country
colname1 = db.financial_aid
colname2 = db.year1
colname3 = db.equipmentloss1
```

```python
df = pd.read_csv('country.csv')
df2 = pd.read_csv('financialaid.csv')
df3 = pd.read_csv('year1.csv')
df4 = pd.read_csv('equipmentloss1.csv')
```

```python
data = df.to_dict(orient = "records")
data1 = df2.to_dict(orient = "records")
data2 = df3.to_dict(orient = "records")
data3 = df4.to_dict(orient = "records")
```

```python
colname.insert_many(data)
colname1.insert_many(data1)
colname2.insert_many(data2)
colname3.insert_many(data3)
```

# Most supportive Country to help Ukraine

```python
a = list(db.financial_aid.aggregate([
    { "$project": 
     {
             '_id': "$Country",
            'totalsum': {'$add': ['$Financial','$Humanitarian','$Military'],},
             'sort': { 'total': -1 },
         
        }
    }
]))

pprint(a)
```

# Total Number of Refugees from Ukraine 

```python
b = list(db.country.aggregate([
    { "$group": 
     {
             '_id': "$Country",
            'No_of_Refugees': { '$max': "$No_of_Refugees" }}},
            {'$sort': { 'No_of_Refugees': -1} }       
    
]))


pprint(b)
```

# Loss of Equipments classified by Air,Land and Water.

```python
a = list(db.equipmentloss1.aggregate([
    { "$project": 
     {
             '_id': "$date",
            'air': {'$add': ['$aircraft','$helicopter','$drone'],},
             'land': { '$add': ['$tank','$field_artillery','$APC','$field_artillery','$MRL','$military_auto','$Anti-aircraft warfare',] },
             'water': {'$add' : ['$naval_ship','$cruise_missiles']},
             'sort': { 'date': -1 },
         
        }
    }
]))

pprint(a)
```
